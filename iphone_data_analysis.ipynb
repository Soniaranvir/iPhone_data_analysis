```python
# Extracting the color from the 'Product Name' (assuming color is mentioned in the name)
# This assumes that the color is mentioned before the storage capacity (e.g., 'Midnight Green', 'Gold')
color_counts = df['Product Name'].str.extract(r'(\w+ \w+)')[0].value_counts()

print("Number of products for each color variant:")
print(color_counts)
```

```python
#Distribution of Star Ratings
#This will help understand the spread of customer ratings for iPhones.
plt.figure(figsize=(8, 6))
sns.histplot(df['Star Rating'], bins=10, kde=True, color='blue')
plt.title('Distribution of Star Ratings')
plt.xlabel('Star Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
#Scatter Plot: Discount Percentage vs. Number of Ratings
#This visualization helps to understand if higher discounts attract more customer reviews.

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Discount Percentage', y='Number Of Ratings', data=df, hue='Star Rating', palette='viridis')
plt.title('Discount Percentage vs. Number of Ratings')
plt.xlabel('Discount Percentage')
plt.ylabel('Number of Ratings')
plt.legend(title='Star Rating', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
#Price Trends Across Models
#A bar chart to show the average sale price of different iPhone models.

avg_price_by_model = df.groupby('Model Name')['Sale Price'].mean().sort_values()

plt.figure(figsize=(10, 8))
avg_price_by_model.plot(kind='barh', color='orange')
plt.title('Average Sale Price by iPhone Model')
plt.xlabel('Average Sale Price')
plt.ylabel('iPhone Model')
plt.show()
```

```python
#Relationship Between Sale Price and Star Rating
#A scatter plot to analyze if higher-priced models have better ratings.

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Sale Price', y='Star Rating', data=df, hue='Star Rating', palette='coolwarm')
plt.title('Sale Price vs. Star Rating')
plt.xlabel('Sale Price')
plt.ylabel('Star Rating')
plt.legend(title='Star Rating', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
#Heatmap: Correlations Between Variables
#This will show correlations between numerical variables like prices, discounts, and ratings.

plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='Blues', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
